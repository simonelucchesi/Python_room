Ecco un semplice esempio (calcolo e rappresentazione grafica della funzione seno): 

```python
from math import pi,sin,cos 

import numpy as np 

import matplotlib.pyplot as plt
```

```python
ripetizioni = 100000

salto = (2*pi)/ripetizioni
valore = 0

lista = []
```

```python
#Caso 1 definizione di funzione iterativa ed applicazione

def my_sin(rip): 
    
    lista = []
    valore = 0
    salto = 2*pi/rip
    
    for _ in range(rip+1): 
    
        lista.append(sin(valore))
        valore +=salto 
        
    return lista
    
```

```python
%timeit lis = my_sin(ripetizioni)

print(lista[0:5])
print(lista[-4:])
```

```python
#Caso 2: codice più snello utilizzando list_comprehension 

%timeit new_lista = [sin(salto*n) for n in range(ripetizioni+1)]

print(new_lista[0:5])
print(new_lista[-4:])
```

```python
vettore = np.array(new_lista)

x = np.arange(0,2*pi,salto)
y = vettore

plt.plot(x,y)
```

### 2.2.1) Controlli di flusso <a name="2.2.1"></a>

Introdotto il concetto di valori di tipo Boolean possiamo adesso presentare due _building blocks_ della programmazione: 

- _If statements:_ consentono di eseguire diversi blocchi di codice a certe condizioni. 

- _Loops:_ consentono di ripetere medesime linee di codice fino a che una certa condizione non è incontrata. 

Loops e If statements possono essere tra loro combinati. 

**If statements: if, elif, else** 

```python
x = 5 

if  x%2 ==0: 
    
    print("Il numero {} è pari".format(x))
        
elif x%2 != 0: 
    
    print("Il numero {} è dispari".format(x))
```

Un paio di cose da notare in questo esempio: 

1) Python per isolare stringe di codice utilizza il sistema ad indentazione: come possiamo notare lo statement print(...) è distanziato dal resto, in questo modo qualora la condizione dovesse essere soddisfatta l'interprete entrerà in quella sotto-sezione eseguendola. 

2) L'operatore `%` restituisce il resto e non il quoziente dell'operazione di divisione con dividendo x e divisore 2. Se un numero diviso 2 ha come resto è necessariamente pari. 

3) La funzione `.format` applicata alla stringa è una tecnica in grado di rendere più semplice la formattazione di stringhe.

4) `If` viene inserito al primo posto di una struttura _if statement_ se la condizione passata è **Vera** verrà eseguito il codice indentato. 

5) `Elif` può essere usato per passare condizioni addizionali da valutare all'interno della struttura, anche in quuesto caso se la condizione passata è **Vera** verrà eseguito il codice indentato. 

6) `Else` non richiede condizioni, se le condizioni poste precedentemente non sono **Vere** verrà eseguito il codice a questi indentato. 

**Loops: for e while** 

- **For- in** loop: in un loop di questo tipo definiamo l'intervallo (o un dato insieme) su cui vogliamo generare un loop, una ripetizione e una variabile (di controllo) assumerà tutti i valori compresi nell'intervallo di riferimento. 



```python
for nulla in [1,2,3,4,5]: 
    
    
    print(nulla)
```

In questo caso l'oggetto passato è una _data structure_ chiamata lista, contente numeri interi. Stiamo creando un flusso su una struttura _iterabile_ andando ad ogni step a valutare ogni elemento corrispondente. 

Quanti loop stiamo generando? 

```python
counter = 0 

for n in [0,1,2,3,4]: 
    
    print(n)
    counter+=1 
print("N° di iterazioni: {}".format(counter))
```

Si noti come l'oggetto su cui stiamo eseguendo l'iterazione può essere anche semplicemente utilizzato come supporto per capire quante iterazioni effettuare, svincolandoci completamente dagli elementi in esso contenuti. 



```python
counter = 0 
x = 2 

print(x)
for _ in [1,1,1]: 
    
    x = x**2
    print(x)
    
    counter+=1

print("N° di iterazioni: {}".format(counter))
```

Un'altra funzione molto utile per quando andremo ad eseguire iterazioni è `range()`: essa è in grado di creare strutture iterabili contenenti numeri in successione. 

Nello specifico `range(inizio,fine,step)`:

## 3) Funzioni in Python <a name ="3"></a>

Per funzione, come già visto in qualche esempio, ci riferiamo ad un blocco di codice riutilizzare creato ad-hoc per svolgere determinate operazioni. 

Le funzioni permettono una maggiore modularizzazione del codice amplificandone l'efficienza: infatti, una volta definita una funzione è possibile applicarla su input diversi semplicemente chiamandola, evitando ogni volta di doverne definire di nuove. 


```python
def numeri_pari(lista): 
    
    #Input = lista 
    
    counter = 0 
    
    for elem in lista: 
        if elem%2 ==0:
            counter+=1 
    
    return counter #Output
```

E' possibile anche definire funzioni senza che queste abbiano un _return_ statement: 

```python
def quanti_numeri_dispari(lista): 
    
    counter = 0 
    for elem in lista:
        if elem%2 !=0:
            counter+=1 
            
    print(counter)
    
```

```python
lista = [1,4,3,6,5,1,3]

numeri_pari(lista)
quanti_numeri_dispari(lista)
```

**Esempio: Successione di Fibonacci** 

Una serie di Fibonacci rappresenta una successione numerica in cui ciascun numero è dato dalla somma dei due precedenti: 1,1,2,3,5,8... 

Proviamo a definire una funzione in grado di generare una successione di lunghezza input = n

```python
def Fibonacci(n): 
    
    lis = []
    
    assert n > 0, "La serie non è realizzabile, deve contenere almeno un numero "
    
        
    F_1, F_2 = 1,1
    
    for i in range(n):
        
        
        lis.append(F_1) 
        
        F_skip = F_1 #Variabile temporanea 
        F_1 = F_2 
        F_2 += F_skip
                  
    
    return lis

            
            
```

```python
Fibonacci(5)
```
